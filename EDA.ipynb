# EDA

```python
#missing value check
df.isnull().sum()
```

```python
# Mengisi missing values pada kolom 'bmi' dengan median
df['bmi'].fillna(df['bmi'].median(), inplace=True)
```

```python
#missing value check
df.isnull().sum()
```

```python
# check duplicate
df1 = df.apply(lambda x:sum(x.duplicated()))
print(df1)
```

```python
# Membuat list numerik dengan pengecualian kolom yang disebutkan
nums = [i for i in df.columns if df[i].dtypes != 'object' and i not in ['id', 'hypertension', 'heart_disease', 'ever_married', 'stroke']]

# Menentukan ukuran plot
plt.figure(figsize=(15, 10))

# Membuat boxplot untuk setiap kolom numerik yang dipilih
for i in range(len(nums)):
    plt.subplot(3, 3, i+1)  # Menyesuaikan jumlah baris dan kolom sesuai kebutuhan
    sns.boxplot(y=df[nums[i]], color='green', orient='v')
    plt.title(nums[i])

plt.tight_layout()  # Mengatur layout setelah loop selesai
plt.show()
```

```python
# distribusi data
plt.figure(figsize=(15, 4))
for i in range(0, len(nums)):
    plt.subplot(3, 4, i+1)
    sns.distplot(df[nums[i]], color='green')
    plt.tight_layout()
```

```python
from sklearn.preprocessing import LabelEncoder

# Inisialisasi LabelEncoder
label_encoder = LabelEncoder()

# Kolom yang ingin diubah
categorical_columns = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']

# Mengonversi setiap kolom
for column in categorical_columns:
    df[column] = label_encoder.fit_transform(df[column])

# Menampilkan beberapa data teratas untuk memeriksa hasil
df.head(15)
```

```python
# Correlation for All Coloums
plt.figure(figsize=(7, 7))
sns.heatmap(df.corr(), annot=True, fmt='.2f')
```
