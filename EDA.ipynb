## Data Cleaning

```python
#@title Menghapus kolom 'Outcome' dari DataFrame
drop_col = ['Outcome']
df = df.drop(columns=drop_col)
num_cols = [i for i in df.columns if i not in drop_col]
```

```python
df.info()
```

```python
#@title Handling Outlier

# Fungsi untuk menghitung persentase outlier
def hitung_persentase_outlier(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    batas_bawah = Q1 - 1.5 * IQR
    batas_atas = Q3 + 1.5 * IQR
    jumlah_outlier = ((data < batas_bawah) | (data > batas_atas)).sum()
    persentase_outlier = jumlah_outlier / len(data) * 100
    return persentase_outlier

# Menghitung persentase outlier pada setiap kolom numerik
persentase_outlier = df[num_cols].apply(hitung_persentase_outlier)

print(persentase_outlier)
```

```python
#@title Berapa Banyak Orang yang mempunyai diabetes dan yang tidak punya diabetes?

plt.figure(figsize=(3, 3))
sns.countplot(x='Outcome', data=df)
plt.title('Jumlah Diabates')
plt.show()
```

```python
#@title Distribusi umur

# Membuat salinan DataFrame
df_vis = df.copy()

# Fungsi untuk mengelompokkan umur
def kelompok_umur(umur):
    if 0 <= umur <= 12:
        return 'Anak-Anak'
    elif 13 <= umur <= 19:
        return 'Remaja'
    elif 20 <= umur <= 35:
        return 'Dewasa Muda'
    elif 36 <= umur <= 55:
        return 'Dewasa'
    else:
        return 'Lansia'

# Membuat kolom baru 'kelompok_umur' dengan menerapkan fungsi ke kolom 'umur'
df_vis['kelompok_umur'] = df_vis['Age'].apply(kelompok_umur)

# Membuat bar chart
fig = px.histogram(df_vis, x="kelompok_umur", title='Jumlah Kelompok Jenis Kelamin', width=400, height=400)
fig.show()
```

```python
#@title Kelompok Usia mana yang paling banyak memiliki diabetes?

#membuat variable orang yang terkena diabetes
strok=df_vis.loc[df_vis['Outcome']==1]
#membuat histogram
fig = px.histogram(strok, x="kelompok_umur", title='Jumlah Orang Diabetes berdasarkan Kelompok Umur', width=500, height=400)
fig.show()
```

```python
#@title Hipertensi dengan diabetes

# Definisikan batas-batas kategori tekanan darah
def categorize_blood_pressure(bp):
    if bp < 80:
        return 'Normal'
    elif 80 <= bp < 90:
        return 'Prehypertension'
    else:
        return 'Hypertension'

# Terapkan fungsi untuk mengkategorikan tekanan darah
df['BloodPressure_Category'] = df['BloodPressure'].apply(categorize_blood_pressure)

# Membuat histogram dengan kategori tekanan darah dan diabetes dengan warna yang ditentukan
fig = px.histogram(
    df,
    x='BloodPressure_Category',
    color='Outcome',
    title='Distribusi Kategori Tekanan Darah berdasarkan Diabetes',
    barmode='group',
    width=600,
    height=400,
    color_discrete_map={0: 'blue', 1: 'red'}  # warna biru untuk Outcome = 0 dan merah untuk Outcome = 1
)

fig.show()
```
