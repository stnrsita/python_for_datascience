# Lakukan Filter Dataframe

```python
#@title Filter with logic operator

data_with_tax_more_than_250 = data[data.tax >= 250]
data_with_tax_more_than_250
```

```python
#@title Multiple filter

multiple_filter = data[(data.tax >= 250) & (data.fuelType == 'Petrol') & (data.year > 2010) & (data.transmission == 'Manual')]
multiple_filter
```

```python
#@title Filter with isin

kolom_engineSize= [2.0, 3.0, 4.0]

filter_with_isin = data[data.engineSize.isin(kolom_engineSize)]
filter_with_isin
```

```python
#@title Filter with str accessor

data[data.transmission.str.startswith('Se')]
```

```python
data[data.model.str.endswith('ra')]
```

```python
#@title Filter with nlargest & nsmallest

#Mengambil 8 data belanja pengguna dengan  total pembelian terbanyak
data.nlargest(8, 'price')
```

```python
# Mengambil 8 data belanja pengguna dengan  total pembelian terbanyak
data.nsmallest(8, 'price')
```

```python
#@title Filter with Query

data.query('year > 2010 and transmission == "Manual" and fuelType == "Petrol" and  engineSize >= 1.5')
```

# Lakukan replace dataframe

```python
#@title Replace Values in Column
# replace(oldvalue, newvalue)

data2['transmission'] = data2['transmission'].replace('Manual','Standar')
print(data2)
```

```python
data2['mpg'] = data2['mpg'].replace(36.2, 36.5)
print(data2)
```

```python
data2['year'][1] = 2016
print(data2)
```

```python
#@title Replace multiple values in column

data2['year'] = data2['year'].replace([2011, 2017], 2016)
print(data2)
```

```python
data2['fuelType'] = data2['fuelType'].replace(['Petrol', 'Diesel'], ['Bensin', 'Solar'])
print(data2)
```
