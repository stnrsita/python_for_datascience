# MATPLOTLIB

```python
# Contoh line chart: menunjukkan tren rata-rata nilai emp_var_rate per bulan
# Group by 'month' untuk menghitung rata-rata emp_var_rate per bulan
monthly_emp_rate = df.groupby('month')['emp_var_rate'].mean()

# Membuat line chart
plt.figure(figsize=(10, 6))
plt.plot(monthly_emp_rate.index, monthly_emp_rate.values, marker='o', color='b')

# Memberi judul dan label
plt.title('Average Employment Variation Rate by Month')
plt.xlabel('Month')
plt.ylabel('Average Employment Variation Rate')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Histogram dari variabel umur
plt.figure(figsize=(10, 6))
plt.hist(df['age'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Barplot dari status pekerjaan
plt.figure(figsize=(10, 6))
df['job'].value_counts().plot(kind='bar', color='orange')
plt.title('Barplot of Job')
plt.xlabel('Job')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns

# 1. Countplot variabel 'marital'
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='marital', palette='Set2')
plt.title('Countplot of Marital Status')
plt.show()
```

```python
# 2. Boxplot umur berdasarkan pendidikan
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='education', y='age', palette='Set3')
plt.title('Boxplot of Age by Education')
plt.xticks(rotation=45)
plt.show()
```

```python
import plotly.express as px

# 2. Pie chart distribusi status pernikahan
fig = px.pie(df, names='marital', title='Distribution of Marital Status')
fig.show()
```

```python
# Pie chart for Job Categories
fig = px.pie(df, names='job', title='Job Distribution')
fig.show()
```

```python
#Scatter plot umur vs euribor3m dengan hasil (y)
fig = px.scatter(df, x='age', y='euribor3m', color='y', title='Scatter Plot of Age vs Euribor3m')
fig.show()
```
