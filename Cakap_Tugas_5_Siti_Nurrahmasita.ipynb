# Gambaran Umum Dataset: Data Penjualan Big Mart

---

## Konteks

Dataset ini terdiri dari data penjualan yang dikumpulkan oleh para ilmuwan data di Big Mart untuk tahun 2013. Dataset ini mencakup informasi tentang 1559 produk yang dijual di 10 toko yang berlokasi di berbagai kota. Tujuan utama dari dataset ini adalah untuk membangun model prediktif yang dapat memprediksi penjualan setiap produk di berbagai outlet. Dengan memanfaatkan model ini, Big Mart berharap dapat memperoleh wawasan mengenai atribut penting dari produk dan outlet yang secara signifikan mempengaruhi kinerja penjualan.

## Konten

Dataset ini dibagi menjadi dua bagian:

- **Set Pelatihan**: Berisi 8.523 pengamatan mengenai informasi outlet item beserta nilai penjualannya (**yang saya gunakan**).
- **Set Uji**: Terdiri dari 5.681 kombinasi outlet item untuk yang perlu diprediksi penjualannya.

## Tautan ke Dataset
Anda dapat mengakses dataset ini di [Kaggle](https://www.kaggle.com/datasets/akashdeepkuila/big-mart-sales).

## Deskripsi Variabel

Dataset ini mencakup variabel-variabel berikut:

1. **ProductID**: Pengidentifikasi unik untuk setiap produk.
2. **Weight**: Berat produk (dalam kg).
3. **FatContent**: Menunjukkan apakah produk rendah lemak atau tidak (kategori: 'Low Fat', 'Regular').
4. **Visibility**: Persentase total area display yang dialokasikan untuk produk dalam sebuah toko.
5. **ProductType**: Kategori produk (misalnya, Milk, Snack Foods, dll.).
6. **MRP**: Harga Eceran Maksimum (harga yang tertera) produk.
7. **OutletID**: Pengidentifikasi unik untuk setiap toko.
8. **EstablishmentYear**: Tahun pendirian outlet.
9. **OutletSize**: Ukuran toko yang diukur berdasarkan luas area yang ditempati (kategori: 'Small', 'Medium', 'High').
10. **LocationType**: Jenis kota tempat toko berada (misalnya, Tier 1, Tier 2).
11. **OutletType**: Menentukan format outlet (misalnya, Grocery Store, Supermarket).
12. **OutletSales**: (Variabel Target) Total penjualan produk di toko tersebut.

## Nilai yang Hilang

Penting untuk dicatat bahwa dataset ini mungkin mengandung nilai yang hilang karena beberapa toko tidak melaporkan semua data akibat masalah teknis. Oleh karena itu, pembersihan dan pemrosesan data yang tepat diperlukan untuk menangani nilai-nilai yang hilang ini sebelum membangun model prediktif.

## Inspirasi untuk Analisis

Penjualan produk tertentu di outlet ritel dapat dipengaruhi oleh atribut toko dan karakteristik produk. Dataset ini menjadi dasar yang sangat baik untuk mengeksplorasi berbagai teknik ilmu data, termasuk analisis regresi, pembelajaran mesin, dan rekayasa fitur, untuk memprediksi penjualan di masa depan dengan akurasi yang lebih baik.

# Penerapan Functional Programming

## Lambda

```python
#@title Lambda sederhana

df['DiscountedPrice'] = df['MRP'].apply(lambda x: x * 0.9)
df[['MRP', 'DiscountedPrice']].head()
```

```python
#@title Lambda with list comprehension

df['AdjustedVisibility'] = df.apply(lambda row: row['ProductVisibility'] * row['OutletSales'], axis=1)

# Create 'SalesCategory' using a list comprehension with conditional logic
df['SalesCategory'] = [
    'High' if sales > 200 else 'Medium' if sales > 100 else 'Low'
    for sales in df['AdjustedVisibility']
]

print("\nDataFrame after adding new columns:")
print(df[['ProductID', 'ProductVisibility', 'OutletSales', 'AdjustedVisibility', 'SalesCategory']])
```

```python
#@title Lambda with if-else

df['WeightCategory'] = df['Weight'].apply(lambda x: 'Light' if x < 10 else 'Heavy')

print("DataFrame with WeightCategory:")
df[['Weight', 'WeightCategory']].tail()
```

```python
#@title Lambda with filter

filtered = list(filter(lambda x: x > 450, df['OutletSales']))
print(filtered[:10])
```

```python
#@title Lambda with map

# Penaikan harga dari masing-masing produk
mapped_mrp = list(map(lambda x: x * 1.1, df['MRP']))
print(mapped_mrp[:10])
```

```python
#@title Lambda with reduce

from functools import reduce

total_sales = reduce(lambda x, y: x + y, df['OutletSales'])
print(total_sales)
```

## Recursion

```python
def iterative_sum(sales):
    total = 0
    for sale in sales:
        total += sale
    return total

sales_sum = iterative_sum(df['OutletSales'].tolist())
print(sales_sum)
```

## First Class Function

```python
#@title Function to a variable

def convert_to_grams(weight):
    return weight * 1000  # Convert kg to grams

# Assigning the function to a variable
weight_converter = convert_to_grams

df['Weight_in_grams'] = df['Weight'].apply(weight_converter)

print(df[['Weight', 'Weight_in_grams']].head())
```

```python
#@title Function as an argument

def process_weight(weight):
    return weight * 10  # Convert kg to hm

# Function as an Argument
def apply_function(func, weight):
    return func(weight)

# Menggunakan for loop untuk menerapkan fungsi
processed_weights = []
for weight in df['Weight']:
    processed_weights.append(apply_function(process_weight, weight))

df['Processed_Weight'] = processed_weights

print(df[['Weight', 'Processed_Weight']].head())
```

```python
#@title Returning a function as a value

# Define a function that returns another function to multiply weights
def weight_multiplier(factor):
    def multiply(weight):
        return weight * factor
    return multiply

# Create a function to triple the weight
triple_weight = weight_multiplier(3)

df['Weight_tripled'] = df['Weight'].apply(triple_weight)

print(df[['ProductID', 'Weight', 'Weight_tripled']].head())
```

# NumPy

## Statistic

```python
import numpy as np
```

```python
# Calculate statistics
mean_sales = np.mean(df['OutletSales'])
median_sales = np.median(df['OutletSales'])
std_dev_sales = np.std(df['OutletSales'])
variance_sales = np.var(df['OutletSales'])
min_sales = np.min(df['OutletSales'])
max_sales = np.max(df['OutletSales'])
quartiles = np.percentile(df['OutletSales'], [25, 75])

# Calculate Q1, Q3, and IQR
Q1 = quartiles[0]
Q3 = quartiles[1]
IQR = Q3 - Q1

print("Sales Statistics:")
print("-------------------------")
print(f"Mean                 : {mean_sales:.2f}")
print(f"Median               : {median_sales:.2f}")
print(f"Standard Deviation    : {std_dev_sales:.2f}")
print(f"Variance             : {variance_sales:.2f}")
print(f"Minimum              : {min_sales:.2f}")
print(f"Maximum              : {max_sales:.2f}")
print(f"Quartile Pertama (Q1): {Q1:.2f}")
print(f"Quartile Ketiga (Q3) : {Q3:.2f}")
print(f"Rentang Interkuartil (IQR) : {IQR:.2f}")
```

## Matriks

```python
# Matrix Operations
# 1. Addition
added_values = df['Weight'] + df['ProductVisibility']
print("\nAddition (Weight + ProductVisibility):")
print(added_values.head())

# 2. Subtraction
subtracted_values = df['OutletSales'] - df['MRP']
print("\nSubtraction (OutletSales - MRP):")
print(subtracted_values.head())

# 3. Multiplication
multiplied_values = df['Weight'] * df['MRP']
print("\nMultiplication (Weight * MRP):")
print(multiplied_values.head())

# 4. Division
divided_values = df['OutletSales'] / df['Weight']
print("\nDivision (OutletSales / Weight):")
print(divided_values.head())

# 5. Power
powered_values = np.power(df['ProductVisibility'], 2)
print("\nPower (ProductVisibility ^ 2):")
print(powered_values.head())

# 6. Trigonometric Functions
sine_values = np.sin(df['ProductVisibility'])  # Scale by pi for better visibility
print("\nTrigonometric Function (sin(ProductVisibility * pi)):")
print(sine_values.head())
```

```python
#@title Transpose, reshape, dan flatten matrik

matrix = np.array(df[['MRP', 'OutletSales']].values)

# Transpose the matrix
transposed = matrix.T

# Reshape into a 1D array (number of elements = number of rows * number of columns)
reshaped = matrix.reshape(-1)  # This will automatically determine the correct shape

# Flatten the matrix
flattened = matrix.flatten()

print(f"Original Matrix):\n{matrix[:10]}\n")
print(f"Transposed:\n{transposed}\n")
print(f"Reshaped:\n{reshaped[:10]}\n")
print(f"Flattened:\n{flattened[:10]}\n")
```

# Visualization dengan Matplotlib dan Seaborn

## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
#@title Line plot untuk tren penjualan berdasarkan tahun
plt.figure(figsize=(12, 6))

# Menghitung dan menggambarkan total penjualan per tahun
total_sales_per_year = df.groupby('EstablishmentYear')['OutletSales'].sum()
plt.plot(total_sales_per_year.index, total_sales_per_year.values, marker='o', color='orange', linewidth=2)

# Menambahkan judul dan label
plt.title('Tren Total Penjualan Berdasarkan Tahun', fontsize=18)
plt.xlabel('Tahun', fontsize=14)
plt.ylabel('Total Outlet Sales', fontsize=14)

# Pengaturan sumbu dan grid
plt.xticks(total_sales_per_year.index)
plt.grid(visible=True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

## Seaborn

```python
import seaborn as sns
```

```python
#@title Bar plot untuk rata-rata penjualan berdasarkan tipe produk
plt.figure(figsize=(10, 6))

# Hitung rata-rata penjualan berdasarkan tipe produk
average_sales = df.groupby('ProductType')['OutletSales'].mean().sort_values(ascending=False)

# Buat palet warna baru yang lebih menarik, misalnya 'viridis'
colors = sns.color_palette('viridis', len(average_sales))

# Buat bar plot dengan y sebagai tipe produk dan x sebagai rata-rata penjualan
sns.barplot(x=average_sales, y=average_sales.index, palette=colors)

# Tambahkan judul dan label
plt.title('Rata-rata Penjualan Berdasarkan Tipe Produk', fontsize=16)
plt.xlabel('Rata-rata Outlet Sales', fontsize=14)
plt.ylabel('Tipe Produk', fontsize=14)

plt.show()
```
